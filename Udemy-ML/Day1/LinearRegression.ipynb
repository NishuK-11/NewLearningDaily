```python
import pandas as pd
```

```python
from sklearn.datasets import fetch_california_housing
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
df=fetch_california_housing()
```

```python
df
```

```python
dataset=pd.DataFrame(df.data)
```

```python
dataset
```

```python
dataset.columns = df.feature_names
```

```python
dataset
```

```python
#independent and dependent features
dataset.head()
```

```python
x=dataset
y=df.target
```

```python
y
```

```python
#train test
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x,y,test_size=0.33,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
##standarizing the dataset
# StandardScaler values ko aise badal deta hai ki sab features ek jaisi scale par aa jaayein.
# Isse Machine Learning model sahi tarah kaam karta hai, biased nahi hota.
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
x_train=scaler.fit_transform(x_train)
#xnew​=(x−mean)/std
# | Value | Formula      | Scaled Value |
# | ----- | ------------ | ------------ |
# | 10    | (10−20)/8.16 | **−1.22**    |
# | 20    | (20−20)/8.16 | **0**        |
# | 30    | (30−20)/8.16 | **+1.22**    |
```

```python
x_test=scaler.transform(x_test)
# Yaha mean aur std fir se calculate nahi honge!
# Hum wahi mean(20) aur std(8.16) use karenge jo training se mile the.
```

```python
x_train
```

```python
#scaler.inverse_transform(x_train)
```

```python
from sklearn.linear_model import LinearRegression
#cross validation
from sklearn.model_selection import cross_val_score
# Main tumhara test 5 baar 5 different topics se loongi aur akhir me tumhara final result in 5 marks ka average hoga.
```

```python
regression=LinearRegression()
regression.fit(x_train,y_train)
```

```python
mse = cross_val_score(regression,x_train,y_train,scoring='neg_mean_squared_error',cv=5)
```

```python
np.mean(mse)
```

```python
#prediction
reg_predict=regression.predict(x_test)
```

```python
reg_predict
```

```python
import seaborn as sns
sns.displot(reg_predict-y_test,kind='kde')
```

```python
from sklearn.metrics import r2_score
```

```python
score=r2_score(reg_predict,y_test)
```
